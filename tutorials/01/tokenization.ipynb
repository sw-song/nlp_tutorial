## Step 1. Import Tokenizer

````python
from konlpy.tag import Okt

okt = Okt()
````

## Step 2. Load Dataset

````python
from konlpy.corpus import kolaw, kobill

# 헌법 말뭉치 목록
print(kolaw.fileids())

# 국회 법안 말뭉치 목록
print(kobill.fileids())
````

````python
# 앞에서 1000글자 추출
x_kolaw = kolaw.open('constitution.txt').read()
x_kolaw[:1000]
````

````python
x_kobill = kobill.open('1809898.txt').read()
x_kobill[:1000]
````

## Step 3. Tokenization with Okt(Twitter)

````python
print('=====================================')
print(f'형태소 추출\n: {okt.morphs(x_kolaw[:50])}')
print('=====================================')
print(f'명사 추출\n: {okt.nouns(x_kolaw[:50])}')
print('=====================================')
print(f'품사 태깅\n: {okt.pos(x_kolaw[:50])}')
````

## step 4. Visualization

### 4-1. Word Frequancy

````python
from nltk import Text # 토큰 탐색 모듈

nouns_kolaw = okt.nouns(x_kolaw)
nouns_graph = Text(nouns_kolaw) 
nouns_graph.vocab()
````

````python
import matplotlib.pyplot as plt
from matplotlib import rc
plt.style.use('seaborn')
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(12,4))
nouns_graph.plot(50)# 상위 50개 명사 토큰 빈도 시각화
plt.show()
````

### 4-2. WordCloud

````python
from wordcloud import WordCloud

font_path = '/System/Library/Fonts/Supplemental/AppleGothic.ttf'

wc = WordCloud(width = 1000, height = 600, background_color="white", font_path=font_path)
plt.imshow(wc.generate_from_frequencies(nouns_graph.vocab()))
plt.axis("off")
plt.show()
````
